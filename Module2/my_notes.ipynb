# Jupyter Notes

Jupyter notebooks use the `.ipynb` extension
#### Jupyter shortcuts:
- Ctrl + Enter: Runs current cell
- Shift + Enter: runs current cell and opens new cell
- `Esc` then `y` : Makes cell `code` mode
- `Esc` then `m` : Makes cell `markdown` mode

One of the most magic commands is the `matplotlib inline` command. This command will enable plots to be displayed directly inside the notebook

```python
%matplotlib inline
```

```python
"""Simple demo of scatter plot """
import numpy as np
import matplotlib.pyplot as plt

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
#Calculate area
area = np.pi * (15*np.random.rand(n))**2 # 0 to 15 point radius
#Plot data
plt.scatter(x,y, s=area, c=colors, alpha=0.5)
plt.show()
```
